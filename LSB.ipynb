```python
from LSB import Steganography
from evaluation import pixel_errors
from PIL import Image
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_DIR = 'data/kodak'
```

```python
img1_dir,img2_dir = np.random.choice(os.listdir(DATA_DIR), 2, replace=False)
```

```python
fig = plt.figure(figsize=(12,12))
img1 = Image.open(os.path.join(DATA_DIR, img1_dir))
img2 = Image.open(os.path.join(DATA_DIR, img2_dir))
fig.add_subplot(1,2,1)
plt.imshow(img1)
plt.title('Carrier Image')
fig.add_subplot(1,2,2)
plt.imshow(img2)
plt.title('Secret Image')
plt.show()
```

```python
merged_image = Steganography.merge(img1, img2)
plt.imshow(merged_image)
plt.title('Covered Image')
plt.show()
```

```python
unmerged_image = Steganography.unmerge(merged_image)
plt.imshow(unmerged_image)
plt.title('Restored secret Image')
plt.show()
```
